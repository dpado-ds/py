# Расчет количества врачей по данным портала "Доктор на работе"

```python
from bs4 import BeautifulSoup as bs
import urllib.request
```

```python
url = urllib.request.urlopen("https://www.doktornarabote.ru/Home/DetailedStatistic/").read()
soup = bs(url, "lxml")
```

```python
table1 = soup.find_all('table')[0]
```

```python
verify = []
```

```python
for i in table1.find_all('td', {'align' : 'right'}):
    verify.append(int(i.text))
```

```python
sum(verify)
```

врачей с разбивкой по специальностям.

# Поиск ВУЗов с наиб. числом студентов по данным "Медкампус"

```python
url1 = urllib.request.urlopen("http://www.medcampus.ru/Home/DetailedStatistic").read()
soup1 = bs(url1, "lxml")
```

```python
table2 = soup1.find_all('table')[0]
```

```python
import pandas as pd
```

```python
data = table2.findAll('td', { 'width' : '380'})
```

```python
names = [i.text.strip('\n') for i in data]
```

```python
data1 = table2.findAll('td', { 'align' : 'right'})
```

```python
numbers = [int(i.text) for i in data1]
```

```python
dataframe = pd.DataFrame(index = names)
dataframe["# of students"] = numbers
```

Выведем первую 20-ку ВУЗов по числу студентов

```python
top25 = dataframe.sort_values('# of students', ascending = False)[0:25]
top25
```

В список попали 2 ВУЗа не из России.

```python
top23 = top25.drop(['Национальный медицинский университет им. А.А. Богомольца (НМУ), Киев', 
            'Белорусский государственный медицинский университет (БГМУ), Минск'])
top23
```

Остались 23 российских ВУЗа с наибольшим количеством студентов

# Лучшие медвузы по рейтингу "ТОП-100" журнала "Эксперт РА"

```python
url2 = urllib.request.urlopen("http://raexpert.ru/rankings/vuz/vuz_2016/#3").read()
soup2 = bs(url2, "lxml").findAll('table', class_='ranking')[0].findAll('tr')[1:]
```

```python
meds = []
for uni in soup2:
        try:
            if 'мед' in uni.findAll('td', class_='ranking_content')[2].text:
                meds.append(uni)
        except IndexError:
            pass
```

```python
len(meds)
names1 = [uni.findAll('td', class_='ranking_content')[2].text for uni in meds]
ratings0 = [uni.findAll('td', class_='ranking_content')[0].text for uni in meds]
ratings2 = [uni.findAll('td', class_='ranking_content')[5].text for uni in meds]
```

В рейтинге 16 медвузов (медвузов в название которых не входит слова "медицинский", не было)

```python
dataframe1 = pd.DataFrame(index = names1, columns = ['Рейтинг', 'Востребованность у работодателя'])
```

```python
dataframe1['Рейтинг'] = ratings0
dataframe1['Востребованность у работодателя'] = ratings2
```

```python
dataframe1
```

# Средний балл ЕГЭ поступивших (качество абитуриентов):

ВУЗа имеющих наиболее кол-во студентов (зарегистрированных на Медкампусе) не вошли в ТОП-100. Мы уже можем сказать, что усилия по
рекрутингу молодых сотрудников стоит сосредоточить на выпускниках ВУЗов, попавших в оба списка, так как это позволит с большей уверенностью судить и о качестве подготовки и о возможности того, что новый сотрудник посоветует DOC+ своим знакомым по ВУЗу.
```python
driver.get('http://postyplenie.ru/rates/detail.php?ELEMENT_ID=297')
```

```python
unis3 = bs(driver.page_source, 'lxml').find('table', class_ = "simple_table reting").findAll('tr')
unis3 = unis3[2:]
```

```python
marks = pd.DataFrame(index = [i.findAll('td')[0].text for i in unis3], columns = ['Имя', 'Сред. балл'])
```

```python
marks['Имя'] = [i.findAll('td')[1].text for i in unis3]
marks['Сред. балл'] = [i.findAll('td')[2].text for i in unis3]
```

```python
marks.index = marks['Имя']
```

Объединение одних и тех же ВУЗов под разными именами:

```python
from fuzzywuzzy import fuzz
```

```python
matches = {}
for i in top23.index:
    try:
        matches[i] = sorted([j for j in dataframe1.index if fuzz.token_sort_ratio(i,j) >= 65], key=lambda x: fuzz.token_sort_ratio(x, i))[-1]
    except IndexError:
        pass
matches
```

```python
matches = {key: matches[key] for key in matches if key not in ['Кубанский государственный медицинский университет (КубГМУ), Краснодар',
                                                               'Новосибирский государственный медицинский университет (НГМУ), Новосибирск',
                                                              'Ростовский государственный медицинский университет (РостГМУ), Ростов-на-Дону']}
matches
```

```python
top23.index = [matches[i] if i in matches else i for i in top23.index]
```

```python
new = pd.concat([top23, dataframe1], axis = 1, join = 'outer')
```

```python
new
```

На большом массиве такой грубый алгоритм производит много выбросов

```python
matches1 = {}
for i in new.index:
    try:
        matches1[i] = sorted([j for j in marks.index if fuzz.token_sort_ratio(i,j) >= 65],
                             key=lambda x: fuzz.token_sort_ratio(x, i))[-1]
    except IndexError:
        pass
matches1
```

```python
matches1 = {key: matches1[key] for key in matches1 if key not in ['Омский государственный медицинский университет Министерства здравоохранения РФ',
                                                               'Сибирский государственный медицинский университет Министерства здравоохранения РФ',
                                                              'Тюменский государственный медицинский университет Министерства здравоохранения РФ',
                                                              'Уральский государственный медицинский университет Министерства здравоохранения РФ']}
matches1
```

```python
new.index = [matches1[i] if i in matches1 else i for i in new.index]
```

```python
new_new = pd.concat([new, marks], axis = 1, join_axes = [new.index])
```

```python
new_new
```
